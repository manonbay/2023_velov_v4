```python
start_time = time.time()

def create_geojson_features(df):
    features = []
    
    for i in range(len(df)):
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point', 
                'coordinates': [df['longitude'].iloc[i], df['latitude'].iloc[i]]
            },
            'properties': {
                'time': df["horodate"].iloc[i],
                'style': {'color' : ''},
                'icon': 'circle',
                'iconstyle':{
                    'fillColor': df['couleurpardefaut'].iloc[i],
                    'fillOpacity': 0.4,
                    'stroke': 'true',
                    'radius': df['availabilities.all.types'].iloc[i] # dependant de la quantité de velov en station
                }
            }
        }
        features.append(feature)
    return features
df_geojson = create_geojson_features(premierfevrier)

print(f"Time taken: {(time.time()-start_time)}")
```
